```python
import seaborn as sms
import plotly.graph_objects as go
```

```python
# 1.
tit=sms.load_dataset('titanic')
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tit['age'],y=tit['fare']))
fig.show()
```

```python
# 2
fig=go.Figure()
fig.add_trace(go.Box(x=tit['age'],y=tit['fare']))
fig.show()
```

```python
# 3.
tips=sms.load_dataset('tips')
```

```python
tips
```

```python
import plotly.express as px

fig = px.bar( x=tips['sex'], y=tips['total_bill'], color=tips['smoker'], pattern_shape=tips['day'])
fig.show()
```

```python
# 4
i=sms.load_dataset('iris')
import plotly.express as px

fig = px.bar( x=i['sepal_length'], y=i['sepal_width'], color=i['species'])
fig.show()
```
