```python
2.
Gradient descent is an optimization algorithm used to minimize the cost function of a machine learning model.

Batch Gradient Descent: This variant computes the gradient of the cost function with respect to all training examples in each iteration. It is guaranteed to converge to the global minimum of the cost function, but it can be slow and memory-intensive for large datasets.
Stochastic Gradient Descent (SGD): This variant computes the gradient of the cost function with respect to a single training example in each iteration. It is faster and less memory-intensive than batch gradient descent, but it can be noisy and may not converge to the global minimum.
Mini-batch Gradient Descent: This variant computes the gradient of the cost function with respect to a small subset of training examples in each iteration. It combines the advantages of batch and stochastic gradient descent, striking a balance between convergence speed and memory requirements.
Momentum-based Gradient Descent: This variant uses a momentum term to accelerate convergence in directions with consistent gradients and dampen oscillations in directions with high variance gradients. It can converge faster than vanilla SGD, but it requires tuning of additional hyperparameters.
Adaptive Gradient Descent: This variant adapts the learning rate for each parameter based on its historical gradients. It can converge faster than vanilla SGD and requires less tuning than momentum-based gradient descent, but it can be computationally expensive.
```

```python
7.RMSprop optimizer is an optimization algorithm that uses an adaptive learning rate to address the challenges of traditional gradient descent optimization methods. It computes the moving average of the squared gradient to scale the learning rate for each parameter. This helps the optimizer converge faster than vanilla SGD and requires less tuning than momentum-based gradient descent.

RMSprop is similar to Adam optimizer in that it uses an adaptive learning rate. However, RMSprop does not use momentum, which can make it more stable and less sensitive to hyperparameters. On the other hand, Adam optimizer combines the advantages of momentum-based and adaptive gradient descent, making it more computationally efficient and well-suited for large datasets and complex models.

In terms of convergence speed, Adam optimizer is generally faster than RMSprop optimizer. However, RMSprop optimizer is more stable and less sensitive to hyperparameters than Adam optimizer. The choice of optimizer depends on several factors such as dataset size, model complexity, and available computational resources.
```

```python
8.
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
mninst=tf.keras.datasets.mnist
```

```python
(x_train_full,y_train_full),(x_test,y_test)=mninst.load_data()
x_valid,x_train=x_train_full[:5000]/255,x_train_full[5000:]/255
y_valid,y_train=y_train_full[:5000],y_train_full[5000:]
plt.imshow(x_train[0],cmap="binary")
plt.show()
```

```python
x_train[0].shape
```

```python
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras import Sequential
import pandas as pd
```

```python
def optimizer_cheak(op,ep=20):
  model=Sequential()
  model.add(Flatten(input_shape=(28,28)))
  model.add(Dense(100,activation="relu"))
  model.add(Dense(25,activation="relu"))
  model.add(Dense(10,activation="softmax"))
  model.compile(optimizer=op,metrics=["accuracy"],loss='sparse_categorical_crossentropy')
  his=model.fit(x_train,y_train,validation_data=(x_valid,y_valid),batch_size=500,epochs=ep)
  print(op)
  pd.DataFrame(his.history).plot()
```

```python
optimizer_cheak("SGD")
```

```python
from keras.optimizers import SGD
optimizer_cheak(SGD( momentum=0.9))
```

```python
optimizer_cheak("Adagrad",ep=20)
```

```python
optimizer_cheak("Adam",ep=20)
```

```python
optimizer_cheak("RMSprop",ep=20)
```
