```python
import bokeh.io
import bokeh.plotting
import seaborn
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
```

```python
data=seaborn.load_dataset('iris')
```

```python
1.
output_file('welcome.html')
f=figure(title='samble',toolbar_location=None)
f.circle(data['petal_length'],data['petal_width'])
show(f)
#first we create file to shore html file 
# as noraml we create figure with title as objest
# using obj we call requide graphs 
# finally we call show 


```

```python
2.
# Glyphs are the building blocks of Bokeh visualizations.
output_file('welcome.html')
f=figure(title='samble',toolbar_location=None)
f.line(data['petal_length'],data['petal_width'])
show(f)
```

```python
# 3.
# we can change appelba
# by createing funstion we can send paramnet to obj

output_file('welcome.html',)
f=figure(title='samble',toolbar_location=None,x_axis_label='x',y_axis_label='y')
f.line(data['petal_length'],data['petal_width'])
show(f)
```
