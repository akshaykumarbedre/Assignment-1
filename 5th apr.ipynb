```python

1.
import pandas as pd 
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
2.
df.isnull().sum()
```

```python
df.describe()
```

```python
li=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"]
for i in range(len(li)):
    df[li[i]]=df[li[i]].replace(0,(df[li[i]].mean()))

```

```python
df.describe()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df['Outcome']
y
```

```python
3.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
para={
    "criterion":["gini", "entropy", "log_loss"],
    "splitter":["best", "random"],
    "max_depth":[1,2,3,4,5,6,7],


}
```

```python
4. 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
clf = DecisionTreeClassifier()
gscv=GridSearchCV(clf,param_grid=para,cv=5)
```

```python
gscv.fit(x_train, y_train)
```

```python
gscv.best_params_
```

```python
y_pred=gscv.predict(x_test)
y_pred
```

```python
5.
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test ))
print(classification_report(y_pred,y_test ))
```

```python
clf1 = DecisionTreeClassifier(max_depth=4,criterion="entropy")
clf1.fit(x_train, y_train)
y_pred=clf1.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
6.
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(30,20))
tree.plot_tree(clf1, filled=True)
```

```python
7.
clf1 = DecisionTreeClassifier(max_depth=4)
clf1.fit(x, y)
y_pred=clf1.predict(x_test)
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
plt.figure(figsize=(30,20))
tree.plot_tree(clf1, filled=True)
# when we use text data , then predict not validation 
```
