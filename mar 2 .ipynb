```python
2.
# Scatter plot is plot in mathplotlib data is visualization used  based on cardinate point
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scactter plot")
plt.scatter(x,y)

```
